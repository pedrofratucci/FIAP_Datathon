```python
# Train DF
train_df = pd.DataFrame()

for x in range(1, 7):
    train_path = f"data/challenge-webmedia-e-globo-2023/files/treino/treino_parte{x}.csv"
    df = pd.read_csv(train_path)
    train_df = pd.concat([train_df, df])

train_df.tail()   
```

```python
# Itens
itens_df = pd.DataFrame()

for x in range(1,4):
    item_path = f"data/challenge-webmedia-e-globo-2023/itens/itens/itens-parte{x}.csv"
    df = pd.read_csv(item_path)
    itens_df = pd.concat([itens_df,df])
    
itens_df
```

```python
# Check if TimeStampHistory and timestampHistory_new equal columns
time_stamp = train_df[["timestampHistory","timestampHistory_new"]]
time_stamp[time_stamp["timestampHistory"] != time_stamp["timestampHistory_new"]]
```

```python

# Divide os timestamps por vírgula em uma nova coluna
split_timestamps = time_stamp["timestampHistory"].str.split(',')

# Achata os timestamps e converte para numpy array
flattened = np.hstack(split_timestamps.values)

# Converte os timestamps para datetime (vetorizado)
converted = pd.to_datetime(flattened.astype(np.int64) // 1000, unit='s')

# Reconstrói o DataFrame agrupando as datas por linha original
time_stamp["convertedTimestamps"] = split_timestamps.apply(
    lambda row: [datetime.fromtimestamp(int(ts.strip()) / 1000).strftime('%Y-%m-%d %H:%M:%S') for ts in row]
)

# Exibe o resultado
time_stamp.head()
```

```python
# Explode a coluna `convertedTimestamps` para obter cada timestamp individual
all_timestamps = time_stamp["convertedTimestamps"].explode()

# Converta as strings de volta para datetime para garantir a correta ordenação
all_timestamps = pd.to_datetime(all_timestamps)

# Calcula o menor e o maior timestamp
oldest_news = all_timestamps.min()
newest_news = all_timestamps.max()

# Exibe os resultados
print("Notícia mais antiga:", oldest_news)
print("Notícia mais nova:", newest_news)
```

```python
# Ligando history com page para obter mais infos da navegação do usuário

# Cria uma cópia para não modificar o original
df_treino_exploded = train_df.copy()

# 1) Splita os IDs por vírgula -> vira lista
# 2) Explode cada lista de IDs em múltiplas linhas
df_treino_exploded['history'] = df_treino_exploded['history'].str.split(',')

df_treino_exploded = df_treino_exploded.explode('history')

# Remover espaços em branco (caso existam)
df_treino_exploded['history'] = df_treino_exploded['history'].str.strip()
```

```python
df_treino_exploded.reset_index(drop=True)
df_merged = df_treino_exploded.merge(
    itens_df,
    left_on='history',   # Coluna no df de treino (explodido)
    right_on='page',     # Coluna no df de itens
    how='left'           # Ou 'inner', dependendo se quer descartar itens que não existem no df_itens
)
```
